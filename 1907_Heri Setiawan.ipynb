```python
for i in range(len(df.columns)): 
    if (df[df[i].isnull()].shape[0]>0): 
        print('\nAttribute-',i,' (before) :',df[df[i].isnull()].shape[0]) 
        if (df[i].dtypes in ['int64','float64']): 
            df[i].fillna(df[df[i].notnull()][i].mean(), inplace=True) 
        else: 
            df[i].fillna(df[df[i].notnull()][i].mode(), inplace=True) 
        print('\nAttribute-',i,' (after) :',df[df[i].isnull()].shape[0])

df.describe()
```

# Feature Extraction

isLuxury is a new feature that will be true if a car has front-engine (attribute 8 is front) and rear-wheel drive (attribute 7 is rwd). 

```python
df['isLuxury'] = df.apply(lambda row: 1 if ((row[7]=='rwd') and (row[8]=='front')) else 0, axis=1)
```

```python
df['isLuxury'] = df['isLuxury'].astype("category") 
df.head()
```
